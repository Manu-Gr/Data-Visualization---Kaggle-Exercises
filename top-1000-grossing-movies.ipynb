```python
movie_filepath = "../input/top-1000-highest-grossing-movies/Highest Holywood Grossing Movies.csv"
```

```python
plt.figure(figsize=(20,15))
sns.barplot(x=movie_data['World Sales (in $)'],y=movie_data['Genre'])
plt.show()
```

```python
plt.figure(figsize=(20,15))
sns.barplot(x=movie_data['World Sales (in $)'],y=movie_data['Distributor'])
plt.show()
```

We can see that Distributors Like New Market Films, Walt Disney Studio Motion Pictures ,Summit Entertainment have produced highest world sales

```python
plt.figure(figsize=(20,15))
sns.barplot(x=movie_data['International Sales (in $)'],y=movie_data['Distributor'])
plt.show()
```

We can see that Distributors Like  Walt Disney Studio Motion Pictures ,Summit Entertainment and Dream Works have produced highest International sales

```python
plt.figure(figsize=(20,15))
sns.barplot(x=movie_data['Domestic Sales (in $)'],y=movie_data['Distributor'])
plt.show()
```

We can see that Distributors Like New Market Films,IFC Films,Summit Entertainment and Walt Disney Studio Motion Pictures  have produced highest Domestic sales

```python
plt.figure(figsize=(20,20))
sns.pairplot(data= movie_data)
plt.show()
```
